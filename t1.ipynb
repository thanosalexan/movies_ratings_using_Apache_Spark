As first steps in Task1, we created a SparkSession using the SparkSession.builder API with application name 'task1'. The getOrCreate() method is called to either get an existing SparkSession with the same application name or create a new one if it doesn't exist. So we created a new one.

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('q2').getOrCreate()
```

Next we used the read.json() method to read the JSON file that contains our dataset and load it into a DataFrame in Spark. The dataframe is being cached to memory, allowing Spark to access it quickly without having to read it from the JSON file again. The cache() does not immediately cache the data, but instead marks it for caching. The data will be cached in memory the first time an action operation (e.g., show(), count(), etc.) is performed on the DataFrame. Also we displayed the first 20 rows of the dataset.

```python
df = spark.read.json('movie.json', multiLine=True).cache()
```

```python
df.show(20)
```

We created a view 'movies' of our existing dataframe, in order  be used to run SQL-like queries on the data in the DataFrame.

```python
df.createOrReplaceTempView("movies")
```

Next we make an SQL query in order to find the “title” and “year” of the movie with the largest “users_rating” that its title starts with the first letter of my last name (A). In SELECT we chose title, year and in WHERE we used the LIKE in order to isolate only the movies starting with 'a' or 'A'. Next we ordered them in descending order based on the users_rating value,  and we used LIMIT 1 to select the one with the highest users_rating. We used show() to display the outcome.

```python
spark.sql("""
  SELECT  title, year
  FROM movies
  WHERE title LIKE 'a%' OR title LIKE 'A%'
  ORDER BY users_rating DESC
  LIMIT 1
  """).show()
```

Next we make an SQL query in order to find the average “users_rating” of the movies that their title starts with the second letter of my last name (L). In SELECT we chose the average of users_rating and in WHERE we used the LIKE in order to isolate only the movies starting with 'l' or 'L'. We used show() to display the outcome.

```python
spark.sql("""
  SELECT  AVG(users_rating)
  FROM movies
  WHERE title LIKE 'l%' OR title LIKE 'L%'
  """).show()
```

Finally, we make an SQL query in order to find the “title” and “year” of the movie with the most votes that its title starts with the third letter of my last name (E). In SELECT we chose title, year and in WHERE we used the LIKE in order to isolate only the movies starting with 'e' or 'E'. Next we ordered them in descending order based on the votes value,  and we used LIMIT 1 to select the one with the most votes. We used show() to display the outcome.

```python
spark.sql("""
  SELECT  title, year
  FROM movies
  WHERE title LIKE 'e%' OR title LIKE 'E%'
  ORDER BY votes DESC
  LIMIT 1
  """).show()
```
